# Research Questions

1. What factor leads to customer satisfaction fpr an Airline?
2. Does Passenger class determines customers satisfaction? 

# Research Objectives

The Objectives of the study are to:

1. investigate the relationship between customers satisfcation and gender
2. determine if theres any association between passengers class and satisfaction
3. investigate the association among Age, Flight distance, Departure and Arrival delay
4. test if theres difference in the mean age of male and female customers 
5. test if arrival delay and departure delay are statistically significance
6. make a prediction equation for predicting arrival time with respect to departure time in minutes
7. visualize all the factors and compare with customers satisfaction 

# Research Hypothesis

The Null Hpothesis for this research are:

1. H0:  There is no significance relationship between cuatomer satisfaction and gender

2.  H0:  There is no association between passengers class and satisfaction

3.  H0:  there is no association among all the scale factor

4.  H0:  there is no significnace difference between the mean age of male and female customers

5.  H0:  there is no significance relationship between arrival delay and departure delay

This dataset contains an airline passenger satisfaction survey. What factors are highly correlated to a satisfied (or dissatisfied) passenger? Can you predict 
passenger satisfaction?






## **About the dataset**

**There is the following information about the passengers of some airline:**

1. **Gender:** male or female
2. **Customer type:** regular or non-regular airline customer
3. **Age:** the actual age of the passenger

10. **Gate location:** level of satisfaction with the gate location (0: not rated; 1-5)
11. **Food and drink:** food and drink satisfaction level (0: not rated; 1-5)
12. **Online boarding:** satisfaction level with online boarding (0: not rated; 1-5)
13. **Seat comfort:** seat satisfaction level (0: not rated; 1-5)
14. **Inflight entertainment:** satisfaction with inflight entertainment (0: not rated; 1-5)
15. **On-board service:** level of satisfaction with on-board service (0: not rated; 1-5)
16. **Leg room service**: level of satisfaction with leg room service (0: not rated; 1-5)
17. **Baggage handling:** level of satisfaction with baggage handling (0: not rated; 1-5)
18. **Checkin service:** level of satisfaction with checkin service (0: not rated; 1-5)
19. **Inflight service:** level of satisfaction with inflight service (0: not rated; 1-5)
20. **Cleanliness:** level of satisfaction with cleanliness (0: not rated; 1-5)
21. **Departure delay in minutes**
22. **Arrival delay in minutes**

This data set contains a survey on <b>air passenger satisfaction</b>. The following <b>classification problem</b> is set:

It is necessary to predict which of the <b>two</b> levels of satisfaction with the airline the passenger belongs to:

### REGRESSION ANALYSIS

```python
from sklearn.linear_model import LinearRegression

#### inputting your data set(X and Y)
X = data['Arrival Delay in Minutes']
print("X: ", X)

Y = data['Departure Delay in Minutes']
print("Y: ", Y)


# arranging the data in numpy
x = np.array(X).reshape((-1,1))
y = np.array(Y)
print("here is your data in numpy!")
print(x)
print(y)



```

```python
print("These are the estimatees of the parameters you will need to compute your regression analysis")
# fitting regression model
model = LinearRegression()
model.fit(x, y)
model = LinearRegression().fit(x, y)

# to find the coeficient of determination
r_sq = model.score(x, y)
print('coefficient of determination:', r_sq)

# to find the intercept
A = model.intercept_
print('intercept:', A)

# to find the slope
B = model.coef_
print('slope:', B)

##### Printing equation
print("Prediction Equation :")
print("Y", "=", A, "+", B, "x")
```

### CHI-SQUARE ANALYSIS

```python
chisqt = pd.crosstab(data.Gender, data.satisfaction, margins = True)
print(chisqt)
```

```python
value = np.array([chisqt.iloc[0][0:5].values, chisqt.iloc[1][0:5].values])
print(chi2_contingency(value)[0:3])
```

### Interpretation

From the chi2 analysis above, it is revealed that there is a significance relationship between the passengers gender and Satisfaction at 95% confidence interval, with the p-value (0.0) less than 0.05.
